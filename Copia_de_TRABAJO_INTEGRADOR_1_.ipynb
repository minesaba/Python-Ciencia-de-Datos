```python
# Regresión lineal

# Voy a comparar dos variables: los "precios constantes de las ventas" y el "uso de efectivo".

from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

y = data['ventas_precios_constantes']
x = data['efectivo']

# Crear y ajustar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(x.values.reshape(-1, 1), y)

# Valores de tiempo para la predicción
efectivo_prediccion = np.array([11, 13, 16]).reshape(-1, 1)

# Realizar predicciones
ventas_prediccion = modelo.predict(efectivo_prediccion)

# Graficar los datos reales y la regresión lineal
plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x, modelo.predict(x.values.reshape(-1, 1)), color='red', linewidth=2, label='Regresión Lineal')
plt.xlabel('Pago con efectivo')
plt.ylabel('Ventas (precios constantes)')
plt.legend()
plt.title('Regresión Lineal para predecir el uso de efectivo (como medio de pago)')
plt.show()
```

**Conclusión**: De este gráfico puede concluirse que, a medida que los precios aumentan, los pagos en efectivo disminuyen, de una forma sutil, debido a que los consumidores no suelen portar tanto efectivo y, entonces, cuando el costo es mayor, eligen algún medio de pago electrónico.
